# **4. Cálculo del Valor Presente Neto (VPN)**

image.png

```python
def calcular_vpn(flujos, tasa_descuento):
    return npf.npv(tasa_descuento, flujos)

def calcular_tir(flujos):
    return npf.irr(flujos)
```

```python
tasa_descuento = 0.1

vpns = []
tirs = []

for i in range(df.shape[0]):
    flujos = df.iloc[i, 1:].values
    vpns.append(calcular_vpn(flujos, tasa_descuento))
    tirs.append(calcular_tir(flujos))

df['VPN'] = vpns
df['TIR'] = tirs
df[['VPN', 'TIR']]
```

```python
df.columns = df.columns.str.replace('–', '_')  # Reemplaza guion largo Unicode por guion bajo
df.columns = df.columns.str.replace('A\x96o_', 'Año_', regex=False)       # Elimina espacios adicionales
```

```python
print(df.columns.tolist())
```

```python
#VPN de cada proyecto
# Convertir columnas numéricas a tipo float (eliminar caracteres no numéricos si los hay)
for col in df.columns[1:]:
    df[col] = df[col].replace('[^\d\.-]', '', regex=True).astype(float)

# Tasa de descuento fija
r = 0.10

# Función para calcular VPN usando la fórmula tradicional
def calcular_vpn(row, tasa):
    vpn = 0
    num_flujos = len([col for col in df.columns if col.startswith('Año_')])
    for t in range(0, num_flujos + 1):  # Incluye t = 0
        if t == 0:
            flujo = row['Inversion_Inicial']
        else:
            col_name = f'Año_{t}'
            flujo = row[col_name] if col_name in row else 0
        valor_presente = flujo / ((1 + tasa) ** t)
        vpn += valor_presente
    return vpn

# Aplicar cálculo a cada fila
df['VPN'] = df.apply(lambda row: calcular_vpn(row, r), axis=1)

# Mostrar resultados VPN de cada proyecto
print(df[['Proyecto', 'VPN']])
```

# **5. Cálculo de la Tasa Interna de Retorno (TIR)**

image.png

```python
#Libreria Numpy financial
!pip install numpy-financial
```

```python
import numpy_financial as npf

# Función para obtener lista de flujos de caja por proyecto
def obtener_flujos(row):
    flujos = [row['Inversion_Inicial']]  # Año 0
    for t in range(1, 11):  # Años 1 a 10
        col_name = f'Año_{t}'
        flujos.append(row[col_name] if col_name in row else 0)
    return flujos

# Función para calcular TIR
def calcular_tir(row):
    flujos = obtener_flujos(row)
    return npf.irr(flujos)

# Aplicar TIR a cada fila del DataFrame
df['TIR'] = df.apply(calcular_tir, axis=1)

# Mostrar resultados
print(df[['Proyecto', 'TIR']])
```

# **6. Visualización de resultados**

```python
# Grafica con MATPLOTLIB de VPN
import matplotlib.pyplot as plt

# Ordenar por VPN para una mejor visualización
df_sorted = df.sort_values(by='VPN', ascending=False)

plt.figure(figsize=(14, 8)) # Ajustar tamaño para mejor legibilidad

# Crear el gráfico de barras
bars = plt.bar(df_sorted['Proyecto'], df_sorted['VPN'], color='pink')

# Añadir títulos y etiquetas
plt.title('Valor Presente Neto (VPN) por Proyecto', fontsize=16)
plt.xlabel('Proyecto', fontsize=12)
plt.ylabel('VPN (en unidades monetarias)', fontsize=12)

# Rotar las etiquetas del eje X para que no se solapen
plt.xticks(rotation=75, ha="right", fontsize=10) # ha="right" alinea el texto correctamente

# Añadir una cuadrícula para facilitar la lectura de los valores
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar los valores encima de cada barra
for bar in bars:
    yval = bar.get_height()
    # Formatear el número para que sea más legible (con comas como separadores de miles)
    # y ajustar la posición del texto si el valor es negativo
    if yval >= 0:
        plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:,.0f}', va='bottom', ha='center', fontsize=9)
    else:
        plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:,.0f}', va='top', ha='center', fontsize=9)


# Ajustar el layout para que todo quepa
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Grafica con MATPLOTLIB de TIR
import matplotlib.ticker as mtick

# Ordenar por TIR para una mejor visualización (opcional, pero recomendado)
df_sorted_tir = df.sort_values(by='TIR', ascending=False)

# Crear la figura y los ejes para el gráfico
plt.figure(figsize=(15, 9)) # Ajustar tamaño para mejor legibilidad con muchos proyectos

# Crear el gráfico de barras
# Usamos un color diferente para distinguir del gráfico de VPN si se generasen juntos
bars_tir = plt.bar(df_sorted_tir['Proyecto'], df_sorted_tir['TIR'], color='mediumpurple')

# Añadir títulos y etiquetas descriptivas
plt.title('Tasa Interna de Retorno (TIR) por Proyecto', fontsize=18, pad=20)
plt.xlabel('Proyecto', fontsize=14, labelpad=15)
plt.ylabel('TIR (%)', fontsize=14, labelpad=15)

# Formatear el eje Y para mostrar los valores como porcentajes
# xmax=1.0 indica que un valor de 1.0 en los datos representa 100%
# decimals=1 muestra un decimal en el porcentaje (ej. 15.5%)
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0, decimals=1))

# Rotar las etiquetas del eje X para que no se solapen y mejorar legibilidad
plt.xticks(rotation=70, ha="right", fontsize=10) # 'ha' es horizontal alignment
plt.yticks(fontsize=10)

# Añadir una cuadrícula horizontal para facilitar la lectura de los valores de la TIR
plt.grid(True, linestyle='--', alpha=0.6, axis='y') # 'axis='y'' para solo líneas horizontales

# Mostrar los valores de la TIR encima de cada barra
for bar in bars_tir:
    yval = bar.get_height()
    # Formatear el número como porcentaje con un decimal
    # Ajustar la posición vertical del texto (va)
    # y la posición horizontal (ha)
    plt.text(
        bar.get_x() + bar.get_width() / 2.0,
        yval,
        f'{yval:.1%}', # Formato de porcentaje con 1 decimal
        va='bottom' if yval >= 0 else 'top', # Colocar encima para positivos, debajo para negativos
        ha='center',
        fontsize=8,
        color='black' # Color del texto de la etiqueta
    )

# Ajustar el layout para que todos los elementos (títulos, etiquetas) quepan bien
plt.tight_layout(pad=1.5) # Añadir un poco de padding

# Mostrar el gráfico
plt.show()
```

#¿Cómo se relacionan el VPN y la TIR con la toma de decisiones sobre proyectos de inversión?

Proyectos con VPN positivo (865 de 1000):
Estos proyectos generan más valor del que cuesta la inversión, lo que significa que recuperan la inversión inicial y producen una ganancia adicional descontada al valor presente.

Decisión: Son económicamente viables y deberían considerarse para ejecución.

Proyectos con TIR mayor a la tasa de descuento (también 865 proyectos):
Una TIR superior a la tasa de descuento (por ejemplo, 10%) indica que el rendimiento del proyecto es mayor que el mínimo requerido por los inversionistas o el costo de capital.

Decisión: Son atractivos financieramente, ya que superan el rendimiento mínimo esperado.

Coincidencia entre VPN positivo y TIR superior al costo de capital:
En este caso, ambos criterios coinciden en aprobar los mismos proyectos, lo que refuerza la confiabilidad y solidez de la decisión de inversión.

Relación y utilidad en la toma de decisiones:
VPN indica cuánto valor neto genera el proyecto hoy, expresado en dinero. Es útil para priorizar proyectos según su aporte económico real.

TIR representa la rentabilidad porcentual del proyecto. Es útil para comparar proyectos entre sí, especialmente cuando el presupuesto es limitado o los tamaños de inversión son distintos.

Uso conjunto: Cuando ambos son favorables, como en los 865 proyectos con VPN > 0 y TIR > 10%, se tiene mayor certeza de rentabilidad y sostenibilidad del proyecto.

VPN y TIR son indicadores clave para tomar decisiones de inversión. Su coincidencia en la evaluación de los proyectos estudiados demuestra que ambos permiten identificar opciones rentables, pero el VPN debe ser el criterio principal cuando hay que elegir entre proyectos mutuamente excluyentes.
